```python
# Import SQLAlchemy `automap` and other dependencies here
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, MetaData, Table, Column, ForeignKey
```

```python
#Create Engine, extract employee database
engine = create_engine('postgresql://postgres:nialor1!@localhost:5432/employee_db')
```

```python
# Use the Inspector to explore the database and print the table names
inspector = inspect(engine)
inspector.get_table_names()
```

```python
connection = engine.connect()
result = connection.execute('SELECT * from salaries LIMIT 3;')
for row in result:
    print(row)
```

```python
# Use Inspector to view titles data
columns = inspector.get_columns('salaries')
for column in columns:
    print(column['name'], column['type'])
```

```python
# Reflect Database into ORM class

Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
Base.metadata.create_all(engine)
Salaries = Base.classes.salaries
Titles = Base.classes.titles
```

```python
# Start a session to query the database
session = Session(engine)
```

```python
# Query Salary & Title

salary_results = session.query(Salaries.emp_no, Salaries.salary)
title_results = session.query(Titles.emp_no, Titles.title)
```

```python
#Create Salary, Title dataframes, perform inner join
salary_df = pd.DataFrame(list(salary_results), columns =['emp_no', 'salary']) 
title_df = pd.DataFrame(list(title_results), columns =['emp_no', 'title'] )

#Merge
df_merge = pd.merge(salary_df, title_df, on='emp_no', how='inner').drop('emp_no',axis=1)

dfmax = df_merge.groupby('title').max()
dfmean = df_merge.groupby('title').mean()
dfmin = df_merge.groupby('title').min()
```

```python
dfmax.plot.barh(title='maximum salary by title')
```

```python
dfmean.plot.barh(title='mean salary by title')
```

```python
dfmin.plot.barh(title='minimum salary by title')
```
